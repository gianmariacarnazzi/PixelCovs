```python
def done(img, name):
    
    if not type(img)==str:
        raise Exception('Input image name must be a string')
    
    count = 0
    
    for i in range(-180, 180):
        
        rgb_img = cv2.imread(img)
        hsv = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2HSV)
        hsv = np.array(hsv, dtype=np.float64) 
        hsv[:, :, 0] = hsv[:, :, 0] + i
        hsv[:, :, 0][hsv[:, :, 0] > 255] = 255
        hsv[:, :, 0][hsv[:, :, 0] < 0] = 0
        hsv = np.array(hsv, dtype=np.uint8) 
        res = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
        res = cv2.resize(res, (1080, 1080))
        bgr_img = cv2.imwrite('/Users/gianmariacarnazzi/Desktop/PixelVirus/SamuraiHue/'+ '/#'+ name + str(count) +'.jpeg', res)
        count += 1
        
    return "Done"
```

```python
def lighter(color, percent):
    '''assumes color is rgb between (0, 0, 0) and (255, 255, 255)'''
    color = np.asarray(color)
    white = np.array((255, 255, 255))
    vector = white-color
    res = np.round(color + vector * percent)
    return tuple(e for e in res) 
```
